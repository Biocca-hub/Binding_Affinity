```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import networkx as nx
import itertools
```

```python
file_path ='/content/drive/MyDrive/MASTER/TESI/dati/pair_sl.tsv'
try:
    grafo = pd.read_csv(file_path, sep = "\t")
except Exception as e:
    print(f"An error occurred while loading the file: {e}")
```

```python
grafo
```

```python
G = nx.from_pandas_edgelist(grafo, "Uniprot ID 1", "Uniprot ID 2", edge_attr='Occurrences')
rows = []
for i, cc in enumerate(nx.connected_components(G)):
  for node in cc:
    rows.append({"node": node, "CC_ID": i})

cc_df = pd.DataFrame(rows)
print(cc_df.shape)
cc_df
#cc_df.to_csv("connected_components.tsv", sep="\t", index=False)
```

```python
cc_weights = {}
for i, cc_nodes in enumerate(nx.connected_components(G)):
    subgraph = G.subgraph(cc_nodes)
    total_weight = sum(data.get('Occurrences', 0) for u, v, data in subgraph.edges(data=True))
    cc_weights[i] = total_weight

weights = []
i = 0
for id in cc_df['CC_ID']:
  weights.append(cc_weights[id])

cc_df['Weight Sum']=weights
```

```python
nodes_per_cc = cc_df['CC_ID'].value_counts().to_dict()
for i in range(1,112):
  print(i-1, nodes_per_cc[i-1])
```
